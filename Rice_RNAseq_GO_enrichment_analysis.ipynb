# Loading package

# Loading Go_term file

```r
term2gene<- read.csv('IRGSP-1.0.GO_one2one.csv',header=F,sep=",")
term2name <- read.csv('GO_Description.csv',header=F,sep=",")
```

# DEG&WGCNA filtering group

```r
Coregene <- read.csv('20211111_Three_CLusters_Gene_list.csv',header = T,sep=",")
```

# Enrichment analysis

```r
x <- compareCluster(Coregene, fun='enricher',TERM2GENE=term2gene,TERM2NAME=term2name,pvalueCutoff = 0.01, pAdjustMethod = "BH", qvalueCutoff = 0.05)
```

# Dot plot

```r
dotplot(x, showCategory=30,includeAll=TRUE)
```
